```python
def amznextend(test):
    if test.startswith('http://amzn.com/w/'):
        endedx=len(test)
        wishlistid = test[18:endedx]
        result='https://www.amazon.com/gp/registry/wishlist/'+wishlistid+'/ref=top&tag=reddittwin-20'
        return(result)
    elif test.startswith('http://smile.amazon.com'):
        return(test+'/ref=top&tag=reddittwin-20')
    elif test.startswith('https://smile.amazon.com'):
        return(test+'/ref=top&tag=reddittwin-20')
    elif test.startswith('https://www.amazon.com'):
        return (test+'/&tag=reddittwin-20')
    elif test.startswith('http://www.amazon.com'):
        return (test+'/&tag=reddittwin-20')
    else:
        return(test)


        
```
